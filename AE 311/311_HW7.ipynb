```python
# 2d

c = 1.5 #m
v = 25 #m/s
alpha = 5 * np.pi / 180 #rad
theta = np.linspace(0, np.pi, 50) #0 - pi
gamma = 2 * alpha * v * (1 + np.cos(theta)) / np.sin(theta)

plt.plot(theta, gamma, 'g--')
plt.title('Vortex Strength per Unit Span')
plt.xlabel('Unit Span (in rad)')
plt.ylabel('Gamma (m^2/s)')
plt.xlim([0, np.pi])
plt.show()
```
