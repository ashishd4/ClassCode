```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 141   # m^(-1/2)
U1 = 0.24 # m/s
U2 = 2.4  # m/s
print(3)
```

```python
array = np.zeros([100, 100])

u = U2
i = 1
j = 1
v = 0

while i < 10001:
    x = i/100
    y = j/10000
    
    v = ((u*A*y/(x**0.5))**2 + (A/u/4 * ((x/A**2)/(x**1.5))**2))**0.5
    array[j, i] = v
    max = x**0.5 / A
    
    if y > max:
        array[j, i] = u
    i += 1
    if i > 99:
        j += 1
        i = 1
    if j > 99:
        break

plt.figure(figsize = (10, 6))
plt.pcolormesh(array)
plt.colorbar()
plt.xlabel("Distance in X Direction (cm)")
plt.ylabel("Distance in Y Direction (cm)")
plt.title("Color Map of Flow (U = 2.4 m/s)")
```
