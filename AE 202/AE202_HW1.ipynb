```python
import numpy as np
import matplotlib.pyplot as plt

rho_venus_surface = 65 #kg/m^3
rho_earth_surface = 1.22 #kg/m^3
rho_mars_surface = 0.02 #kg/m^3
rho_titan_surface = 6.09 #kg/m^3

scale_venus = 15.9 #km
scale_earth = 8.5 #km
scale_mars = 11.1 #km
scale_titan = 40 #km

def z(H0, rho, rho0):
    return -H0 * np.log(rho/rho0)
```

```python
zmax_venus = z(scale_venus, rho_venus_surface * (10**(-6)), rho_venus_surface)
zmax_earth = z(scale_earth, rho_earth_surface * (10**(-6)), rho_earth_surface)
zmax_mars = z(scale_mars, rho_mars_surface * (10**(-6)), rho_mars_surface)
zmax_titan = z(scale_titan, rho_titan_surface * (10**(-6)), rho_titan_surface)

range_venus = np.linspace(0, zmax_venus)
range_earth = np.linspace(0, zmax_earth)
range_mars = np.linspace(0, zmax_mars)
range_titan = np.linspace(0, zmax_titan)

rho_venus = rho_venus_surface * np.exp(-range_venus/scale_venus)
rho_earth = rho_earth_surface * np.exp(-range_earth/scale_earth)
rho_mars = rho_mars_surface * np.exp(-range_mars/scale_mars)
rho_titan = rho_titan_surface * np.exp(-range_titan/scale_titan)
```

```python
plt.plot(rho_venus, range_venus, 'r--', label = 'Venus')
plt.plot(rho_earth, range_earth, 'g--', label = 'Earth')
plt.plot(rho_mars, range_mars, 'b--', label = 'Mars')
plt.plot(rho_titan, range_titan, 'y--', label = 'Titan')

plt.legend(loc = 1)

plt.title('Atmospheric Density vs. Altitude')
plt.xlabel('Atmospheric Density (in kg/m^3)')
plt.ylabel('Altitude (in km)')
```
