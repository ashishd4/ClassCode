```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def ratio(gamma, M):
    return (1 + (gamma - 1)/2*M**2)**(1/(1 - gamma))

gamma = 1.4
M = np.linspace(0, 1, 100)

plt.plot(M, ratio(gamma, M), color='blue')
plt.xlabel('Mach number')
plt.ylabel('rho/rho0')
plt.title('Density Ratio vs. Mach Number')

plt.show()
```
