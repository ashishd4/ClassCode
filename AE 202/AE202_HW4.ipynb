```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
#1d

r = np.array([-4684.5, 5088.2, 872.2]) #km
v = np.array([-4.0978, -4.5358, 4.4513]) #km/s
mu = 3.986e5 #km^3/s^2
r_mag = la.norm(r)

h = np.cross(r, v)

e = la.norm((np.cross(v, h) - mu*r/r_mag) / mu)

e
```

```python
#2a

altitude = np.arange(200, 40000, 1) #km
r_e = 6371 #km
r = r_e + altitude #km
G = 6.6743e-11 #m^3/(kg*s^2)
m_e = 5.9722e24 #kg
accel = G*m_e/r**2 #m/s^2

plt.plot(altitude, accel, 'r-')
plt.title('Spacecraft Acceleration at Varying Altitudes')
plt.xlabel('Altitude (in km)')
plt.ylabel('Acceleration (in m/s^2)')
plt.show()
```

```python
#2b

altitude = np.arange(200, 40000, 1) #km
r_e = 6371 #km
r = r_e + altitude #km
mu = 3.986e5 #km^3/s^2

velocity = np.sqrt(mu/r) #km/s

plt.plot(altitude, velocity, 'g-')
plt.title('Spacecraft Velocity at Varying Altitudes')
plt.xlabel('Altitude (in km)')
plt.ylabel('Velocity (in km/s)')
plt.show()
```

```python
#2c

altitude = np.arange(200, 40000, 1) #km
r_e = 6371 #km
r = r_e + altitude #km
mu = 3.986e5 #km^3/s^2

period = 2*np.pi*pow(r, 1.5)/np.sqrt(mu)/3600 #hrs

plt.plot(altitude, period, 'b-')
plt.title('Spacecraft Period at Varying Altitudes')
plt.xlabel('Altitude (in km)')
plt.ylabel('Period (in hrs)')
plt.show()
```

```python
#4d - SEE DOC FOR DISCUSSION

r_min = 6371 + 400 #km
r = np.arange(r_min, 1e6, 1) #km
V_inf = 1.5 #km/s
mu = 3.986e5 #km^3/s^2

velocity = np.sqrt(pow(V_inf, 2) + 2*mu/r) #km/s

plt.plot(r, velocity, 'b-')
plt.title('Spacecraft Velocity vs Orbital Radius')
plt.xlabel('Orbital Radius (in km)')
plt.ylabel('Velocity (in km/s)')
plt.show()
```
